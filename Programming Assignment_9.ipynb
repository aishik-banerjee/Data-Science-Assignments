```python
Write a Python program to check if the given number is a Disarium Number?
```

```python
num=input("Enter the number")
length=len(num)
number=int(num)
dis=0
store=number
while (number!=0):
    n=number%10
    dis=dis+ (n**length)
    number=number//10
    length=length-1
if (dis==store):
    print ("Disarium Number")
else:
    print ("Not a Disarium Number")
```

```python
Write a Python program to print all disarium numbers between 1 to 100?
```

```python
def disarium(num):
    length=len(num)
    number=int(num)
    dis=0
    store=number
    while (number!=0):
        n=number%10
        dis=dis+ (n**length)
        number=number//10
        length=length-1
    if (dis==store):
        return True
    else:
        return False

for i in range(1,101):
    if (disarium(str(i))==True):
        print (f"{i} is a Disarium Number")   
```

```python
Write a Python program to check if the given number is Happy Number?
```

```python
def numSquareSum(num):
    dis=0
    n=0
    while (num!=0):
        n=num%10
        dis=dis+ (n**2)
        num=num//10
    return dis

happy=int(input("Enter the number"))
result=happy
while (result!=1 and result!=4):
    result = numSquareSum(result)  
if (result==1):
    print ("Happy Number")
else:
    print ("Not a Happy Number")
```

```python
Write a Python program to determine whether the given number is a Harshad Number?
```

```python
num=int(input("Enter the number"))
def SumNum(num):
    dis=0
    while (num!=0):
        n=num%10
        dis=dis+n
        num=num//10
    return dis
if (num % SumNum(num) == 0):
    print ("Harshad Number")
else:
    print ("Not a Harshad Number")
```

```python
Write a Python program to print all pronic numbers between 1 and 100?
```

```python
def pronic(num):
    i=1
    while(i<num):
        if (i*(i+1) == num):
            return True
        i=i+1
    else:
        return False
    
for i in range(1,101):
    if (pronic(i)==True):
        print (f"{i} is a Pronic Number")   
    
```
