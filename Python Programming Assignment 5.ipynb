```python
1.	Write a Python Program to Find LCM?
```

```python
a=int(input("Enter The 1st Number: "))
b=int(input("Enter The 2nd Number: "))
greater,lcm=0,0
if(a>b):
    greater=a
else:
    greater=b
while(True):
    if((greater%a==0) and (greater%b==0)):
        lcm=greater
        break
    greater+=1
print(f"LCM : {lcm}")
```

```python
2.	Write a Python Program to Find HCF?
```

```python
a=int(input("Enter The 1st Number: "))
b=int(input("Enter The 2nd Number: "))
if(a<b):
    smaller=a
    greater=b
else:
    smaller=b
    greater=a
while((greater % smaller)!=0):
    smaller, greater = greater%smaller,smaller
print (F"HCF : {smaller}")
```

```python
3.	Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?
```

```python
dec=int(input("Enter The Number: "))
print(f"{dec} in Decimal is {bin(dec)} in Binary")
print(f"{dec} in Decimal is {oct(dec)} in Octal")
print(f"{dec} in Decimal is {hex(dec)} in Hexadecimal")
```

```python
5.	Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?
```

```python
a=int(input("Enter The 1st Number: "))
b=int(input("Enter The 2nd Number: "))
operation=input("Enter 'add' for addition, 'sub' for subtraction, 'mul' for multiplication and 'div' for division")
if (operation=='add'):
    print(a+b)
elif (operation=='sub'):
    print(a-b)
elif (operation=='mul'):
    print(a*b)
elif (operation=='div'):
    print(a//b)
else:
    print ("Enter valid operation")
```
